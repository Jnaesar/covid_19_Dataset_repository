Drop unuse column from covid_19 Datasets :-

```python
covid_df.drop(["Sno", "ConfirmedIndianNational", "ConfirmedForeignNational"], inplace=True, axis=1)
```

```python
#Check Required columns :-
covid_df.head()
```

```python
#Change the format of date column :-

covid_df["Date"]=pd.to_datetime(covid_df["Date"], format="%Y-%m-%d")
covid_df.head()
```

```python
#Find Active cases :-Confirmed cases -(Cured + deaths) each  of the state

covid_df["Active_Cases"] = covid_df["Confirmed"]-(covid_df["Cured"]+covid_df["Deaths"] )
covid_df.tail(10)
```

```python
#Create pivot table
state_wise = pd.pivot_table(covid_df, values=["Confirmed", "Deaths", "Cured"], index="State/UnionTerritory", aggfunc=max)

#Find out recovery rate:- total number of cured Cases*100/confirmed cases
state_wise["Recovery_rate"] = state_wise["Cured"]*100/state_wise["Confirmed"]

#Find out the mortality rate:- total number of deatge *100/total number of confirmed cases:
state_wise["Mortality"] = state_wise["Deaths"]*100/state_wise["Confirmed"]

#Sort values based on the confirmed cases columns:-
state_wise = state_wise.sort_values(by="Confirmed", ascending=False)

state_wise.style.background_gradient(cmap="cubehelix")
```

```python
# Find Top ten active  cases states:-
top_ten_active_cases_state = covid_df.groupby(by="State/UnionTerritory").max()[["Active_Cases", "Date"]].sort_values(by=["Active_Cases"],
ascending=False).reset_index()

plt.figure(figsize=(15, 7))

plt.title("Top ten most active cases states", size=20)

ax = sns.barplot(data=top_ten_active_cases_state.iloc[:10], y="Active_Cases", x="State/UnionTerritory", linewidth=2, edgecolor="black")

plt.xlabel("States")
plt.ylabel("Total active cases")
plt.show()

```

```python
#top ten states with highest deaths :-
top_ten_deaths_states = covid_df.groupby(by="State/UnionTerritory").max()[["Deaths", "Date"]].sort_values(by=["Deaths"], ascending=False).reset_index()

figure=  plt.figure(figsize=(15, 7))

plt.title("Top ten states with highest deaths", size=25)

ax = sns.barplot(data = top_ten_deaths_states.iloc[:12], y="Deaths", x="State/UnionTerritory", linewidth=3, edgecolor="red")
#iloc[:12]=--- because maharastra and karntka has duplicates values

plt.xlabel("states")
plt.ylabel("Total Deaths Caase")
plt.show()
```

```python
#Growth trend running :---
figur = plt.figure(figsize=(15, 7))

ax = sns.lineplot(data=covid_df[covid_df["State/UnionTerritory"].isin(["Maharashtra", "Karnataka", "Kerala", "Tamil Nadu", "uttar Pradesh"]), x="Date", y="Active_Cases", hue="State/UnionTerritory"])
# hue:---Hue is represents the color being displayed, as found on a red-green-blue scale, color wheel or grayscale. 
# The intensity of the primary colors or gray displayed ...

ax.set_title("Top five Affected States in India", size=20)
```

```python
#Drop missing value columns :---
vaccination = vaccine_df.drop(columns=["Sputnik V (Doses Administered)", "AEFI", "18-44 Years (Doses Administered)",
                     "45-60 Years(Individuals Vaccinated)",
                    "60+ Years(Individuals Vaccinated)"], axis=1)
#In pandas axis = 0 refers to horizontal axis or rows and axis = 1 refers to vertical axis or columns.
vaccination.head()

# Finally drop the 5 unused columns:
```

```python
#Remove row where state="India",   (I want to work on only state name rows, because India is not state)

vaccine = vaccine_df[vaccine_df.State!="India"]
vaccine.head()
```

```python
#Rename column name from Total Individuals Vaccinated to total:--

vaccine.rename(columns={"Total Individuals Vaccinated" : "Total"}, inplace=True)
vaccine.head()
```

```python
#Max vaccinated states top 5 :-
max_vac = vaccine.groupby("State") ["Total"].sum().to_frame("Total")
max_vac = max_vac.sort_values("Total", ascending=False).iloc[:5]
max_vac
```

```python
#created bar plot of top 5 vaccinated states:--
fig = plt.figure(figsize=(10, 5))
plt.title("Top five vaccinated states in India", size=20)
x = sns.barplot(data=max_vac.iloc[:10], y=max_vac.Total, x=max_vac.index, linewidth=2, edgecolor="black")
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()


#  reset_index()`will recreate index column every time we run it on same data
```
